```python
import numpy as np

#
# Get the array data from an external CSV data file.
#
array_data = np.genfromtxt(fname='data.csv', delimiter=',')

x = np.array(array_data[0])
y = np.array(array_data[1])


#
# Rewrite the line equation as y = Ap,
# where A = [[x 1]] and p = [[m], [c]].
#
A = np.vstack([x, np.ones(len(x))]).T

#
# Use lstsq to solve for p.
#
m, c = np.linalg.lstsq(A, y, rcond=None)[0]


#
# plot the data along with the fitted line
# 
import matplotlib.pyplot as plt

plt.plot(x, y, 'o',       label='original data', markersize=10)
plt.plot(x, m*x + c, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
#
# pretty printing utilities
#
from IPython.display   import display
from sympy.interactive import printing
printing.init_printing(use_latex=True)


#
# compose a function
#
import sympy as sp

x    = sp.symbols('x', real=True)
fn   = sp.Pow(x, x)

#
# display the function and its derivative
#
display(fn)
display(fn.diff())

#
# plot the function and its derivative
#
p    = sp.plot(
            fn,
            fn.diff(),
            xlim = (0, 5),
            ylim = (-25, 500),
            show = False)

p[0].line_color = 'red'
p[1].line_color = 'green'
p.show()
```
